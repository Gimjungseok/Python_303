```python
t=np.arange(0,12,1)
y1=np.sin(t)
y2=np.cos(t)
plt.plot(t, y1, 'r--', marker='*')
plt.plot(t, y2, 'go-')
plt.grid()
plt.figure(figsize=(12,6))
```

```python
plt.subplot(1,2,1)
plt.plot(t, y1, 'r--', marker='*')
plt.subplot(1,2,2)
plt.plot(t, y2, 'go-')
```

### [문제]

● subplot을 이용해서 1행 2열 구조 2개씩

● 4가지 색상 다르게

● 4가지 스타일 다르게

● 4가지 마커 서로 다르게

● 범례표시

● x,y,제목


```python
x=np.arange(0,5,1)
y1=x+1
y2=x+2
y3=x+3
y4=x+4
```

```python
plt.subplot(1,2,1)
plt.plot(x, y1, 'r--', marker='*')
plt.plot(x, y2, 'b--', marker='+')
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.legend('le')
plt.subplot(1,2,2)
plt.plot(x, y3, 'y--', marker='.')
plt.plot(x, y4, 'g--', marker='^')

plt.xlabel('aing')
plt.ylabel('garosa')
plt.title('asd')
plt.legend('ye')
```
